```python
import os
train_directory="Train2/"
test_directory="Test2/"
```

```python
def images(image_directory):
    return [image_directory+image for image in os.listdir(image_directory)]

def images2(image_directory):
    return [image_directory+image for image in os.listdir(image_directory)]

images(train_directory)
```

```python
##In training directory, image filename indicates image label as cat or dog: Need to extract labels
## Extracting training image labels
train_image_names = images(train_directory)
```

```python
# Function to extract labels
def extract_labels(file_names):
    '''Create labels from file names: class1 = 0 and class2 = 1   and class3 = 2 '''
    
    # Create empty vector of length = no. of files, filled with zeros 
    n = len(file_names)
    y_train = np.zeros(n, dtype = np.int32)
    # Enumerate gives index
    for i, filename in enumerate(file_names):
           
        # If 'class1' string is in file name assign '0'
        if 'class1' in str(filename):
            y_train[i] = 0
        elif 'class2' in str(filename):
            y_train[i]= 1        
        else:
            y_train[i] = 2
    return y_train

extract_labels(train_image_names)
```

```python
# Save labels
y_train = extract_labels(train_image_names)

# Save labels: np.save(file or string, array)
np.savetxt('y_train.txt', y_train)

# Images in test directory
images2(test_directory)
```

```python
##In testing directory, image filename indicates image label as class1 / class2 / class3 : Need to extract labels
## Extracting testing image labels
test_image_names = images2(test_directory)
```

```python
# Function to extract labels
def extract_labels2(file_names):
    '''Create labels from file names: class1 = 0 and class2 = 1   and class3 = 2 '''
    
    # Create empty vector of length = no. of files, filled with zeros 
    n = len(file_names)
    y_test = np.zeros(n, dtype = np.int32)
    # Enumerate gives index
    for i, filename in enumerate(file_names):
           
        # If 'cat' string is in file name assign '0'
        if 'class1' in str(filename):
            y_test[i] = 0
        elif 'class2' in str(filename):
            y_test[i]= 1        
        else:
            y_test[i] = 2
    return y_test

extract_labels2(test_image_names)
```

```python
# Save labels
y_test = extract_labels2(test_image_names)

# Save labels: np.save(file or string, array)
np.savetxt('y_test.txt', y_test)
```

```python
### [2]  Principal Component Analysis

import math
from PIL import Image
```

```python
# Load training and testing data
X_train = np.loadtxt('train_data.txt')
X_test = np.loadtxt('test_data.txt')
print("Shape of training set: {}".format(X_train.shape))
print("Shape of testing set: {}".format(X_test.shape))
```

```python
def image_grid(D,H,W,cols=10,scale=1):
    """ display a grid of images
        H,W: Height and width of the images
        cols: number of columns = number of images in each row
        scale: 1 to fill screen
    """
    n = np.shape(D)[0]
    rows = int(math.ceil((n+0.0)/cols))
    fig = plt.figure(1,figsize=[scale*20.0/H*W,scale*20.0/cols*rows],dpi=100)
    for i in range(n):
        plt.subplot(rows,cols,i+1)
        fig=plt.imshow(np.reshape(D[i,:],[H,W]), cmap = plt.get_cmap("gray"))
        plt.axis('off')
```

```python
# D) PCA
from sklearn.decomposition import PCA
n_components = 40
```

```python
# Extracting the top 40 eigenfaces from

pca = PCA(n_components = n_components,
         svd_solver = 'randomized', whiten = True)
```

```python
# Train your data
pca.fit(X_train)
```

```python
# Showing the eigenfaces
eigenfaces = pca.components_
```

```python
image_grid(eigenfaces[:,:], H, W)
```

```python
plt.show()
```

```python
# E) Projection
# Projecting the input data on the eigen
X_train_pca = pca.transform(X_train)
```

```python
X_test_pca = pca.transform(X_test)
```

```python
X_train_pca.shape
```

```python
X_test_pca.shape
```

```python
# G) Calculate the intra and inter-class distances

#Calculate distances

intra_class_dist = np.sum(np.power((X_train_pca[1,:] - X_test_pca[1,:]), 2))

inter_class_dist = np.sum(np.power((X_train_pca[1,:] - X_test_pca[5,:]), 2))
```

```python
intra_class_dist
```

```python
inter_class_dist
```

```python
# F) Plot the cumulative variance against the number of principal
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```
